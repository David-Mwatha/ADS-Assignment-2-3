```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
data
```

```python
data.head()
```

#  Show a breakdown of distance from home by job role and attrition.

```python
data.columns
```

```python
# display the columns needed and save them to a variable distance
distance= data[['Attrition', 'JobRole', 'DistanceFromHome']]
distance
```

```python
# Group the data by job role and attrition, and calculate the mean distance from home:
mean_distance = distance.groupby(['JobRole', 'Attrition']).mean().reset_index()
mean_distance
```

```python
# Pivot the table to show the breakdown of distance from home by job role and attrition:
distance_pivot = mean_distance.pivot(index='JobRole', columns='Attrition', values='DistanceFromHome')
```

```python
# Plot the data using a bar chart:
distance_pivot.plot(kind='bar', figsize=(10,6))
plt.title('Distance from Home by Job Role and Attrition')
plt.xlabel('Job Role')
plt.ylabel('Distance from Home (in miles)')
plt.show()
```

#  Compare average monthly income by education and attrition

```python

# Group the data by education level and attrition status, and calculate the mean monthly income
grouped_data = data.groupby(['Education', 'Attrition'])['MonthlyIncome'].mean()
grouped_data
```

```python
# Convert the result to a Pandas DataFrame and reset the index
result_data = grouped_data.to_frame().reset_index()
result_data
```
