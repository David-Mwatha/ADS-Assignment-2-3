#  Show a breakdown of distance from home by job role and attrition.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
data.head()
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='JobRole', y='DistanceFromHome', hue='Attrition', data=data, estimator=np.mean)
plt.title('Distance from Home by Job Role and Attrition')
plt.show()

```

# Compare average monthly income by education and attrition.

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Education', y='MonthlyIncome', hue='Attrition', data=data)
plt.title('Monthly Income by Education and Attrition')
sns.set_style('ticks')
plt.show()
```
